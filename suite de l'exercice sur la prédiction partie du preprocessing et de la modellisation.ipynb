### C) Encodage des variables qualitatives
Rappelons que nous avions separez les données qualitatives dans vquali

```python

encoder = OneHotEncoder(sparse=False) # l'option sparse pour afficher dans une liste
enco = encoder.fit_transform(vquali)
encode = pd.DataFrame(enco, index = vquali.index)
encode.head()
```

```python
encode.columns = encoder.get_feature_names(vquali.columns)
encode.head()
```

```python
df1 = pd.concat([encode, vquanti], axis = 1)
df1
```
